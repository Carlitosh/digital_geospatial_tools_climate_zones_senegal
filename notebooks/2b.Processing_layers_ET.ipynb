```python
def paths_to_datetimeindex(paths):
    #return  [datetime.strptime(date.split('/')[-2], '%Y%m%d') for date in paths]
    paths = [os.path.basename(date) for date in paths]  
    return  [datetime.strptime(date.replace('.tiff','').split('y')[1], '%Y%j') for date in paths]
```

### Annually average 8 days 
 1. Create function 
 2. Select each year 
 3. Calculate mean X days
 4. Groupby by days
 5. Apply map to applying function  

```python
def resample_8D_mean(x):
    return x.resample(time="8D").sum('time')
```

```python
ds_grouped_by_days = ds_clipped.groupby('time.year').apply(resample_8D_mean)
```

```python
grouped_yearly_average = ds_grouped_by_days.groupby('time.dayofyear').mean().astype(np.uint8)
#grouped_yearly_average
#grouped_yearly_average['ET_daily'] = grouped_yearly_average['ET_daily'].astype(np.uint8)
grouped_yearly_average
```
