## Download individual images

```python
begin_date = date(1985,7,1)  
end_date = date(2016,12,31)
```

```python
def days_between(begin_date, end_date):
    return abs(end_date - begin_date).days
```

```python
dias_desde =  days_between(end_date, begin_date) 
```

```python
list_files = []
for step in range(-1, dias_desde):
    date = begin_date  + timedelta(days= step+1)
    day = date.strftime('%d')
    month = date.strftime('%m')
    year = date.strftime('%Y')
    query = 'global_tifs_p05/Tmin/{0}/Tmin.{0}.{1}.{2}.tif'.format(year,month,day)
    path = "http://data.chc.ucsb.edu/products/CHIRTSdaily/v1.0/"
    output = r"E:/datawarehouse/CHIRTS_Tmin/"
    list_files.append(str(path+query))
```

```python
len(list_files)
```

```python
def download_url(url):
    print("downloading: ",url)
    file_name = str(url.split('/')[-1])
    output = r"E:/datawarehouse/CHIRTS_Tmin/"
    urlretrieve(url, str(output + file_name)) 
    urlcleanup()
```

```python
# Run 5 multiple threads. Each call will take the next element in urls list
results = ThreadPool(10).imap_unordered(download_url, list_files)
for r in results:
    print(r)
```
